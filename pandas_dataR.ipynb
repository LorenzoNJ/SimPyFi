```python
import pandas as pd
from pandas_datareader import data as web
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,5)
```

```python
# On charge le dataset
stock_list = pd.read_csv('data_finance/stocks.csv')['Stock_Name'].tolist()
index_list = pd.read_csv('data_finance/index.csv')['Index_Symbol'].tolist()
```

```python
tickers = ['MSFT', 'AAPL', 'EURUSD=X']
data = pd.DataFrame()
for t in stock_list:
    data[t] = web.DataReader(t, data_source='yahoo', start='2000-1-1')['Adj Close']
data.plot()
```
